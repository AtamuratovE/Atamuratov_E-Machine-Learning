### 2. Import the Dataset:
    
importing the dataset

```python
salary_data = pd.read_csv('Salary_Data.csv')
X = salary_data.iloc[:, :-1].values
y = salary_data.iloc[:, 1].values
```

### 4. Split the data into training and Testing set:

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

*SO above we are splitting the model , 1/3 is for testing and 2/3 is for training. here y is a y(salary) variable and x is a x(years of experiance) variable*

### 5. Fit Simple Linear Regression to the training dataset:

Fit Simple Linear Regression to the Training set:

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

*in above example we are importing linearRegression model it's alredy ready for us, then we take instance of that to **lr** after which we use **fit** method of LinearRegression to fit model to train our model, then we pass the training data-set **X-independent, y-dependent** then system will learn, and now training complityed*

### 6. Predict the test set results:
    
Predicting the test set results:    

Below we check wether it's predicting correct or not

```python
y_pred = lr.predict(X_test)
y_pred
```

here **predict** is a common method used for predicting

*here predict will only take one parametr which is independent in this case **X-independent** because dependent variable what it will predict*
* if you paid attention you saw that above we passed y_train *it was to train data* but here we don't pass y_test cause model should predict y then we check that comparing to y_test 


### 7.Visualize the train set results: 

Visualizing the Training set results: 

```python
plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.title('Salary ~ Experience (Train set)')
plt.xlabel('Years of Experience')
plt.show()
```

### 8.Visualize the test set results: 

Visualizing the Test set results: 

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### 9. Calculating the residuals:

Finding the Residuals

```python
# Calculating the Residuals
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,y_pred))
print('MSE:', metrics.mean_squared_error(y_test,y_pred))
print('RMSE:', np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```

#### Since the RMSE (Root Mean Square Error) is too less, the model is a good model

What is an error - distance of that point from this line
